```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = plt.imread('C:/Users/vaishnavi/Desktop/DBDA/ML Practical/Day_1/image1.png')
height, width, _ = image.shape
center = (width // 2, height // 2)  
radius = min(center[0], center[1])  
```

```python
Y, X = np.ogrid[:height, :width]
d_c = np.sqrt((X - center[0])**2 + (Y - center[1])**2)
mask = d_c <= radius
m_i = np.zeros_like(image)
m_i[mask] = image[mask]

plt.imshow(m_i)
```
